corr()

```python
import pandas as pd

# Creating a sample DataFrame
data = pd.DataFrame({
    'Math': [80, 85, 90, 70, 95],
    'Science': [75, 70, 85, 65, 90],
    'History': [60, 65, 70, 55, 75]
})

# Calculating the correlation matrix
correlation_matrix = data.corr()

print("Correlation Matrix:")
print(correlation_matrix)
```

In this example, data.corr() calculates the correlation between the 'Math', 'Science', and 'History' columns in the DataFrame data. The resulting correlation matrix shows the correlation coefficients between each pair of columns.

For instance, the correlation between 'Math' and 'Science' is approximately 0.972, indicating a strong positive correlation. Similarly, the correlation between 'Math' and 'History' is approximately 0.993, also indicating a strong positive correlation.

rename columns name

```python
import pandas as pd

def renameColumns(students: pd.DataFrame) -> pd.DataFrame:
    data = students.rename(columns={"id": "student_id", "first": "first_name", "last": "last_name", "age": "age_in_years"})
    return data
```

agr dataset mai null values hai or usse fill krne hai ksi value se wo median, mode, mean of dataset(rows ya columns) kuch bhi hoskta let say null value ki jha 0 place krna hai tu below syntax hoga zero ke ilawa jo bhi place krna chate ho wo likhdo.

```python
import pandas as pd

def fillMissingValues(products: pd.DataFrame) -> pd.DataFrame:
    data = products.fillna(0)
    return data
```

agr specify krna hai ke srf is particular column ki null value ko ksi value se replace krdo tu:

```python
import pandas as pd

def fillMissingValues(products: pd.DataFrame) -> pd.DataFrame:
    values = {"quantity": 0}
    data = products.fillna(value=values)
    return data
```

means just quantity ke column mai null value hai usse 0 se replace kro other ki nhi.or value ko replace krrhe tu value keyword use hoga.

if you want to concatenate 2 dataframes then you can use concat() function of pandas

```python
import pandas as pd

def concatenateTables(df1: pd.DataFrame, df2: pd.DataFrame) -> pd.DataFrame:
    data = pd.concat([df1, df2])
    return data
```

If you want to change the datatype of a column you can you astype() method of pandas

```python
import pandas as pd

def changeDatatype(students: pd.DataFrame) -> pd.DataFrame:
    data = students.astype({'grade':'int32'}) 
    return data
```
