```python
import pandas as pd

# Example dataframes
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'Marks': [85, 90, 78],
    'Name': ['D', 'E', 'F']
})

# Concatenate without sorting columns
df_combined = pd.concat([df1, df2], axis=0, sort=False) #jb 2 df ko concate krte hai tu lexographical order mai sort krta hai means M phele aata hai the N aata hai tu phele Marks aaenge then Name agr hate ho ke phele Name aae then Marks aae tu sort ko False krdo

df_combined
```

```python
import pandas as pd

# Example DataFrames
d1 = pd.DataFrame({
    'Name': ['A', 'B', 'C'],
    'Age': [25, 30, 22]
}, index=[1, 2, 3])

d2 = pd.DataFrame({
    'Marks': [85, 90, 78]
}, index=[2, 3, 4])

# Full Outer Join
result = d1.join(d2, how='outer') #ye sb dedega like 1 index is not present in Marks while 4 is not present in Name,Age wha ye NaN show krkr sb ko print kradega
result
```

```python
import pandas as pd

data = {'Category': ['A', 'A', 'B', 'B'],
        'Subcategory': ['X', 'Y', 'X', 'Y'],
        'Sales': [100, 150, 200, 250]}

df = pd.DataFrame(data)
pivot = pd.pivot_table(df, 
                       values='Sales', 
                       index='Category', 
                       columns='Subcategory', 
                       aggfunc='sum')

pivot
```


A pivot table in pandas is a powerful tool for summarizing data in a DataFrame. It allows you to reorganize and aggregate your data to better understand relationships and patterns. Here's a breakdown of how it works:

Key Features:
Rows and Columns: You can specify which columns from your DataFrame should be used as the rows and columns in the pivot table.
Values: You can choose which data should be aggregated, such as sums, averages, counts, etc.
Aggregation Functions: You can specify how the data should be aggregated, using functions like sum, mean, count, min, max, etc.
Multi-Level Indexing: Pivot tables support multi-level indexing (hierarchical indexing), which allows for more complex data organization.

In this example:

Category becomes the rows.
Subcategory becomes the columns.
Sales are the values that are summed for each combination of Category and Subcategory.
Applications:
Summarizing large datasets.
Transforming data into a more understandable format.
Comparing different subsets of data.
Data analysis and reporting.
This feature is particularly useful in exploratory data analysis and reporting, where you need to quickly summarize and understand data from different angles.
