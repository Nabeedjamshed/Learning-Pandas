```python
#Fill with mode, first find mode then fill it
dataset['Hospital'].mode()[0] #mode per column ka niklta hai
```

```python
dataset['Hospital'].fillna(dataset['Hospital'].mode()[0], inplace=True)
```

```python
dataset
```

```python
# if you want ke just ek column mai nhi sb mai mode lga so you can use loop
for i in dataset.columns:
    dataset[i].fillna(dataset[i].mode()[0], inplace=True)
```

```python
dataset.head(3)
```

```python
# if you want ke just object type ka data fill ho so:
dataset.select_dtypes(include='object').head(3) #numberical data nhiaaega isse
```

```python
for i in dataset.select_dtypes(include='object').columns:
     dataset[i].fillna(dataset[i].mode()[0], inplace=True)
```
